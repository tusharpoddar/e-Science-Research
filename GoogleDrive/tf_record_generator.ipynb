# The following notebook is used to generate the tf-records from the VGGish model. 
In order to plot the cluster plot it is important to extract the tf-records(they provide us with the location of the spectrogram). This file shows how to generate the tf-records for a numpy file that has been generated before.

## What is the use of the tf-records?
The tf-records generated can be used to plot the cluster plot. The tf records help in specifying the relative position of the spectrograms with others. Similar spectrograms are plotted together and the ones that are different are plotted farther away. 

The notebooks does the following - 
1. Asks the user to specify the numpy file that contains all the data that needs to fed to the VGGis model(needs to be in (?, 96, 64)). 
2. Stores the tf-records in the file specified by the user.
